```python
import time
import pandas as pd

user_name = input("Enter the same nickname as what you have been for the previous obstacles: ")

class Game:
    def __init__(self, user_name):
        self.level_number = 1
        self.fail_count = 0
        self.input_amount = 0
        self.user_name = user_name
        self.o_num = 4
        print(f"Welcome, {self.user_name}. Let's learn about variables in Python.")

    def get_question(self, question_number):
        if question_number == 1:
            print("What is the correct way to create a variable equal to your name in Python?")
            print(f"A. name = {self.user_name}")
            print(f"B. My name is: {self.user_name}")
        elif question_number == 2:
            print("How would you set the integer 5 equal to the variable x?")
            print("A. 5")
            print("B. x")
            print("C. x = 5")
        elif question_number == 3:
            print("Marty is a giraffe who is notoriously known for ALWAYS waking up the hyenas at 7:00am with Marty's loud chomping noises. ")
            print("Select the answer that would set the boolean variable wake_up in this scenario.")
            print(" ")
            print("A. wake_up = False")
            print("B. wake_up = marty_true")
            print("C. wake_up = Yes")
            print("D. wake_up = True")

    def play_game(self):
        start_time = time.time()
        while self.level_number <= 2:
            print(f"\nLevel {self.level_number}:")
            if self.level_number == 1:
                self.get_question(1)
                answer = input("Enter your answer (A/B): ")
                self.input_amount += 1
                if answer.upper() == "A":
                    print("Correct! You will now advance to the next level.")
                    self.level_number += 1
                else:
                    self.fail_count += 1
                    print("Incorrect. The correct answer is A.")
                    self.get_question(1)
                    self.level_number -= 1
                    answer = input("Enter your answer (A/B): ")
                    if answer.upper() == "A":
                        print("Correct! You will now advance to the next level.")
                        self.input_amount += 1
                        self.level_number += 2
                    else:
                        self.fail_count += 1
                        print("Incorrect again. The correct answer is A.")
                        self.level_number = 2
                self.level_number += 1
        while self.level_number <= 3:
            self.get_question(2)
            answer = input("Enter your answer (A/B/C): ")
            self.input_amount += 1
            if answer.upper() == "C":
                print("Correct!")
            else:
                self.fail_count += 1
                print("Incorrect. The correct answer is C.")
                self.get_question(2)
                answer = input("Enter your answer (A/B/C): ")
                self.input_amount += 1
                if answer.upper() == "C":
                    print("Correct! You will now advance to the next level.")
                else:
                    self.fail_count += 1
                    print("Incorrect again. The correct answer is C.")
                    self.level_number -= 1
            self.level_number += 1
        while self.level_number == 4:
            self.get_question(3)
            answer = input("Enter your answer (A/B/C/D): ")
            self.input_amount += 1
            if answer.upper() == "D":
                print("Correct!")
            else:
                self.fail_count += 1
                print("Incorrect. The correct answer is D.")
                self.get_question(3)
                answer = input("Enter your answer (A/B/C/D): ")
                self.input_amount += 1
                if answer.upper() == "D":
                    print("Correct! You have completed the game.")
                else:
                    self.fail_count += 1
                    print("Incorrect again. The correct answer is D.")
                    self.level_number -= 1
            self.level_number += 1

                
start_time = time.time()
game = Game(user_name)
game.play_game()
end_time = time.time()
time_spent = end_time - start_time
print(f"\nCongratulations, {user_name}. You have completed the game in {round(time_spent)} seconds.")
print(f"You made {game.fail_count} mistake(s) and entered {game.input_amount} input(s).")
```
